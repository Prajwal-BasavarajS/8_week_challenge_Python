## **Problem 13: Banking Transaction**
This problem combines everything from today's session.

1. Create `balance = 1000.0`
2. Create `withdrawal_amount = "250"` (as a string, simulating user input)

## **Problem 14: Student Enrollment**
1. Create `age = "19"` (as a string)
2. Create `credits_completed = 120`
3. Convert `age` to an integer.
4. Compute two booleans:
   - `eligible_by_age` = `age >= 18`
   - `eligible_by_credits` = `credits_completed >= 100`
5. Create `eligible_to_enroll` = `eligible_by_age` **and** `eligible_by_credits` *(Note: both conditions must be True)*
6. Print all variables clearly:
```
Age:                  19
Credits completed:    120
Eligible by age:      True
Eligible by credits:  True
Eligible to enroll:   True
```

````python
age = "19"
credits_completed = 120
age_int = int(age)
eligible_by_age = bool(age_int >= 18)
eligible_by_credits = bool(credits_completed >= 100)
eligible_to_enroll = bool(eligible_by_age and eligible_by_credits)
print(age_int)
print("eligible_by_age = ", eligible_by_age)
print("eligible_by_credits = ", eligible_by_credits)
print("eligible_to_enroll = ", eligible_to_enroll)
````
