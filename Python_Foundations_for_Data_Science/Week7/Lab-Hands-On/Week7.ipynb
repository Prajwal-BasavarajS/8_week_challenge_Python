# 3. Common Plot Types

### 3.1 Line Plot (Trend Analysis)

Used when:

- Order matters

- Data represents time progression

- We analyse trends

````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May"]
sales = [12000, 13500, 15000, 9000, 20000]

plt.plot(months, sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (€)")
plt.show()
````

### 3.2 Bar Chart (Category Comparison) 

Used when:

- Comparing independent categories

- No inherent order required

````python
import matplotlib.pyplot as plt
departments = ["IT", "HR", "Finance"]
employee_count = [12, 7, 5]

plt.bar(departments, employee_count)
plt.title("Employees per Department")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()
````

### 3.3 Histogram (Distribution Analysis)

Used when:

- Understanding distribution shape

- Checking spread

- Detecting skewness

````python
import numpy as np
import matplotlib.pyplot as plt
# generates 200 random numbers following a bell curve centered at 70 with a spread of 10 (standard deviation), simulating exam scores.
'''In practice it means:
 ~68% of scores will fall within 70 ± 10 (i.e. between 60–80)
 ~95% will fall within 70 ± 20 (i.e. between 50–90)
 Very few scores will land far outside that range'''
scores = np.random.normal(70, 10, 200) 
# takes that data, splits it into 14 buckets, counts how many scores land in each, and draws the bars.
plt.hist(scores, bins=14)
plt.title("Exam Score Distribution")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
````

### 3.4 Scatter Plot (Relationship between variables)

Used when:

- Exploring correlation

- Detecting patterns between two numeric variables

````python
import matplotlib.pyplot as plt
study_hours = [1, 2, 3, 4, 5, 6, 7]
exam_scores = [50, 55, 65, 70, 75, 85, 90]

plt.scatter(study_hours, exam_scores)
plt.title("Study Hours vs Exam Score")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
````

# 4. Working with JSON Data

### 4.1 JSON Structure

JSON supports:

- Objects &rarr; Python dictionaries

- Arrays &rarr; Python lists

- Nested structures

````python
import json

# defines a Python dictionary representing structured company data
company_data = {
    "company": "TechCorp",  # company name
    "location": "Berlin",   # company location
    "employees": [          # list of employees, each represented as a dictionary
        {"name": "Alice", "salary": 4000},  # first employee with name and salary
        {"name": "Bob", "salary": 4500}     # second employee with name and salary
    ]
}

# opens (or creates) a file named "company.json" in write mode
with open("company.json", "w") as file:
    # serialises the Python dictionary into JSON format and writes it to the file
    # indent=4 ensures the JSON output is formatted with readable indentation
    json.dump(company_data, file, indent=4)

# prints a confirmation message after successfully creating the JSON file
print("company.json has just been created!")
````

````python
# reading JSON from file
with open("company.json", "r") as file:
    data = json.load(file)

print(data["employees"][0]["name"])
````

````python
# parsing Nested Data
for employee in data["employees"]:
    print(employee["name"], employee["salary"])
````

### 5.1 What is an API?
**Application Programming Interface** (API)

- An API defines a set of rules that enables different software systems to communicate with each other.
- In web-based systems, APIs are commonly accessed through HTTP requests.

### 6.3 Strategic Print Debugging & Logging
The simplest and most universal debugging tool.

````python
def apply_discount(total):
    if total > 100:
        return total * 0.9
    return total

def process_order(order):
    print(f"[DEBUG] Received order: {order}")

    total = sum(item["price"] for item in order)
    print(f"[DEBUG] Calculated total: {total}")

    discount = apply_discount(total)
    print(f"[DEBUG] After discount: {discount}")

    return discount

def main():
    order = [
        {"name": "Laptop", "price": 800},
        {"name": "Mouse", "price": 20},
        {"name": "Keyboard", "price": 50}
    ]

    final_price = process_order(order)

    print(f"Final price: {final_price}")


if __name__ == "__main__":
    main()
````

````python
import logging  # Imports Python's standard logging module, which provides a flexible framework for emitting log messages.

logging.basicConfig(level=logging.DEBUG)
# Configures the logging system with a minimum severity level of DEBUG.
# This means all messages at DEBUG level and higher
# (INFO, WARNING, ERROR, CRITICAL) will be displayed.


def apply_discount(total):
    if total > 100:
        return total * 0.9
    return total


def process_order(order):
    logging.debug("Received order: %s", order)
    # Logs a debug message showing the input order.
    # The %s is a placeholder replaced with the value of order.

    total = sum(item["price"] for item in order)

    logging.debug("Calculated total: %s", total)
    # Logs another debug message showing the calculated total.

    discount = apply_discount(total)

    logging.debug("After discount: %s", discount)
    # Logs the final value after applying the discount.

    return discount


def main():
    order = [
        {"name": "Laptop", "price": 800},
        {"name": "Mouse", "price": 20},
        {"name": "Keyboard", "price": 50}
    ]

    final_price = process_order(order)

    print("Final price:", final_price)


if __name__ == "__main__":
    main()
````

# 7. Writing Clean & Readable Python Code: PEP 8 Basics
### 7.1 Naming Conventions
```Python
# Variables and functions: snake_case
user_name = "Alice"
def calculate_total_price(price, tax_rate):
    return price * (1 + tax_rate)

# Classes: PascalCase
class ShoppingCart:
    pass

# Constants: UPPER_SNAKE_CASE
MAX_RETRIES = 3
DEFAULT_TIMEOUT = 30

# Avoid
userName = "Alice"        # camelCase
def CalculatePrice():     # PascalCase for a function
    pass
```
---
### 7.2 Indentation & Line Length
```Python
# 4 spaces per indentation level (never tabs)
def greet(name):
    if name:
        print(f"Hello, {name}!")

# Max line length: 79 characters
# Break long lines with parentheses (preferred over backslash)
result = (
    some_long_variable_name
    + another_long_variable_name
    + yet_another_variable
)

# Long function signatures
def create_user(
    username,
    email,
    password,
    is_admin=False,
):
    pass
```
---
### 7.3 Whitespace Rules
```Python
# Spaces around operators
x = 5 + 3
is_valid = x > 0 and x < 10

# No spaces inside brackets
my_list = [1, 2, 3]
my_dict = {"key": "value"}

# One space after commas, not before
def add(a, b):
    return a, b

# Avoid
x=5+3
my_list = [ 1,2,3 ]
def add( a,b ):
    return a,b
```
---
### 7.4 Blank Lines
```Python
# Two blank lines between top-level definitions
class Dog:
    pass


class Cat:           # ← two blank lines before this
    pass


def standalone():    # ← two blank lines before this
    pass


# One blank line between methods inside a class
class Animal:
    def eat(self):
        pass

    def sleep(self):   # ← one blank line
        pass
```
---
### 7.5 Imports
```Python
# One import per line
import os
import sys

# Grouped order: stdlib → third-party → local
import os                  # 1. Standard library
import sys

import requests            # 2. Third-party
import numpy as np

from myapp import utils    # 3. Local modules

# Avoid
import os, sys
from os import *           # Wildcard imports pollute the namespace
```
