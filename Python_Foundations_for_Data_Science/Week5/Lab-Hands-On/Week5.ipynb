## **1. File Systems and Directory Structure**

### Concept
- Files are stored in a hierarchical structure of **directories/folders**.  
- Python can interact with files and directories using the `os` and `pathlib` modules.

````python
# Import the built-in os module
# The os module allows Python to interact with the operating system,
# including working with files, folders, and directory paths.
import os


# -------------------------------------------------------------------
# Step 1: Get the current working directory
# -------------------------------------------------------------------

# os.getcwd() stands for "get current working directory"
# It returns the full path of the directory where the current
# Python script or notebook is running.
cwd = os.getcwd()

# Print the current working directory to the console
print("Current Directory:", cwd)


# -------------------------------------------------------------------
# Step 2: List all files and folders inside the current directory
# -------------------------------------------------------------------

# os.listdir(path) returns a list containing all files
# and subdirectories inside the given path.
#
# Here, we pass the current working directory (cwd)
# so that Python lists everything inside it.
print("Contents:", os.listdir(cwd))


# -------------------------------------------------------------------
# Step 3: Create a new directory (folder)
# -------------------------------------------------------------------

# os.makedirs() creates a new directory.
#
# "data_folder" is the name of the folder to be created.
#
# exist_ok=True prevents Python from raising an error
# if the folder already exists.
#
# Without exist_ok=True:
#     FileExistsError would occur if the folder already exists.
#
# With exist_ok=True:
#     Python safely ignores the request if the folder is already there.
os.makedirs("data_folder", exist_ok=True)

# Print a confirmation message after creating the folder
print("New folder created: data_folder")
````

````python
# -------------------------------------------------------------------
# splitlines(): remove newline characters (\n)
# -------------------------------------------------------------------

# The splitlines() method splits the text into separate lines
# and removes newline characters (\n).
#
# Unlike readlines():
#   - readlines() keeps \n at the end of each line
#   - splitlines() removes \n automatically
#
# This is often cleaner and easier to work with.


# ===================================================================
# Option 1: Traditional file handling
# ===================================================================

print("Option 1: ==================================")


# -------------------------------------------------------------------
# Step 1: Open the file
# -------------------------------------------------------------------

# Open the file in read mode.
read_file = open('reading_file.txt')


# -------------------------------------------------------------------
# Step 2: Read the full file content
# -------------------------------------------------------------------

# .read() reads the entire file as ONE large string.
#
# Example:
# "Welcome to SRH!\nWe are reading file in our "Programming for Data Science: Python Foundations for Data Science" module."
#
# Then, .splitlines() splits the text into multiple lines
# while removing newline characters (\n).
#
# Final result:
# ['Welcome to SRH!', 'We are reading file in our "Programming for Data Science: Python Foundations for Data Science" module.']
text = read_file.read().splitlines()


# -------------------------------------------------------------------
# Step 3: Print the data type
# -------------------------------------------------------------------

# splitlines() returns a list.
#
# Expected output:
# <class 'list'>
print(type(text))


# -------------------------------------------------------------------
# Step 4: Print the processed lines
# -------------------------------------------------------------------

# Each line becomes a separate string inside the list,
# without newline characters.
print(text)


# -------------------------------------------------------------------
# Step 5: Close the file
# -------------------------------------------------------------------

# Always close files after use.
read_file.close()



# ===================================================================
# Option 2: Using with open() (recommended approach)
# ===================================================================

print("Option 2: ==================================")


# -------------------------------------------------------------------
# with open() automatically handles file closing
# -------------------------------------------------------------------

# The with statement is the recommended modern approach
# for file handling in Python.
#
# Why?
#   - It automatically closes the file
#   - It is safer and cleaner
#   - It avoids forgetting close()
with open('reading_file.txt') as r:


    # ---------------------------------------------------------------
    # Step 1: Read and split the text into lines
    # ---------------------------------------------------------------

    # Read the entire file and split it into lines
    # without keeping newline characters.
    text = r.read().splitlines()


    # ---------------------------------------------------------------
    # Step 2: Print the type
    # ---------------------------------------------------------------

    print(type(text))


    # ---------------------------------------------------------------
    # Step 3: Print the resulting list
    # ---------------------------------------------------------------

    print(text)


    # ---------------------------------------------------------------
    # Step 4: Manual close() is unnecessary here
    # ---------------------------------------------------------------

    # This line is technically unnecessary because
    # the with statement automatically closes the file.
    #
    # It is included here only for demonstration purposes.
    r.close()
````

````python
# -------------------------------------------------------------------
# Append a new line to an existing file
# -------------------------------------------------------------------

# The mode "a" stands for append mode.
#
# Append mode:
#   - adds new content to the END of the file
#   - does NOT overwrite existing content
#
# If the file does not exist,
# Python automatically creates it.
with open('reading_file.txt', 'a') as r:


    # ---------------------------------------------------------------
    # Add a newline character
    # ---------------------------------------------------------------

    # '\r' represents a carriage return character.
    #
    # However, in modern systems,
    # '\n' is the standard newline character.
    #
    # Common newline representations:
    #   - '\n'   → Unix/Linux/macOS newline
    #   - '\r\n' → Windows newline
    #   - '\r'   → old Macintosh systems
    #
    # Using only '\r' may not behave consistently
    # across all operating systems.
    #
    # Recommended alternative:
    # r.write('\n')
    r.write('\r')


    # ---------------------------------------------------------------
    # Append new text to the file
    # ---------------------------------------------------------------

    # This line is added at the end of the file.
    r.write('We have just added this line automatically.')


    # ---------------------------------------------------------------
    # Manual close() is unnecessary here
    # ---------------------------------------------------------------

    # Since we are using with open(),
    # the file is automatically closed.
    #
    # Therefore, calling close() manually is redundant.
    r.close()



# -------------------------------------------------------------------
# Read the updated file content
# -------------------------------------------------------------------

# Open the same file again in read mode.
with open('reading_file.txt') as r:


    # ---------------------------------------------------------------
    # Read the entire file and split it into lines
    # ---------------------------------------------------------------

    # .read() reads the full file content as one string.
    #
    # .splitlines() splits the text into separate lines
    # and removes newline characters.
    text = r.read().splitlines()


    # ---------------------------------------------------------------
    # Print the data type
    # ---------------------------------------------------------------

    # splitlines() returns a list.
    #
    # Expected output:
    # <class 'list'>
    print(type(text))


    # ---------------------------------------------------------------
    # Print all file lines
    # ---------------------------------------------------------------

    # The output will be a list where:
    #   - each item represents one line
    #   - newline characters are removed
    print(text)


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    # The with statement automatically closes the file.
    r.close()
````

````python
# -------------------------------------------------------------------
# Writing content into a file
# -------------------------------------------------------------------

# Important note:
# The comment says:
#     "write overwrites any existing content"
#
# However, this code uses mode "a" (append mode),
# NOT mode "w" (write mode).
#
# Difference:
#
# "w" → write mode
#       - overwrites existing file content
#       - creates the file if it does not exist
#
# "a" → append mode
#       - adds content to the end of the file
#       - keeps existing content unchanged
#       - creates the file if it does not exist


# -------------------------------------------------------------------
# Open the file in append mode
# -------------------------------------------------------------------

# If "new_reading_file.txt" does not exist,
# Python automatically creates it.
#
# If it already exists,
# new text will be appended to the end.
with open('new_reading_file.txt', 'a') as r:


    # ---------------------------------------------------------------
    # Write the first line into the file
    # ---------------------------------------------------------------

    # .write() writes text into the file.
    #
    # It returns the number of written characters,
    # although we are not storing that value here.
    r.write('We have just created a new file.')


    # ---------------------------------------------------------------
    # Add a newline character
    # ---------------------------------------------------------------

    # '\n' moves the cursor to the next line.
    #
    # Without '\n',
    # all text would appear on the same line.
    r.write('\n')


    # ---------------------------------------------------------------
    # Write another line into the file
    # ---------------------------------------------------------------

    r.write('A new line has been added.')


    # ---------------------------------------------------------------
    # Add another newline character
    # ---------------------------------------------------------------

    r.write('\n')


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    # Since we are using with open(),
    # the file is automatically closed.
    r.close()



# -------------------------------------------------------------------
# Reading the newly updated file
# -------------------------------------------------------------------

# Open the file again in read mode.
with open('new_reading_file.txt') as r:


    # ---------------------------------------------------------------
    # Read the file and split it into lines
    # ---------------------------------------------------------------

    # .read() reads the full content as one string.
    #
    # .splitlines() splits the content into separate lines
    # and removes newline characters.
    #
    # Final result:
    # a list of strings
    text = r.read().splitlines()


    # ---------------------------------------------------------------
    # Print the file content
    # ---------------------------------------------------------------

    # Example output:
    # ['We have just created a new file.',
    #  'A new line has been added.']
    print(text)


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    r.close()
````

## **3. Working with CSV Files**

### Writing a CSV

```python
import csv

data = [
    ["Name", "Age", "Grade"],
    ["Alice", 20, 85],
    ["Bob", 21, 90],
    ["Charlie", 19, 78]
]

with open("students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

### Reading a CSV

```python
with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

````python
# -------------------------------------------------------------------
# Working with CSV files in Python
# -------------------------------------------------------------------

# CSV stands for:
#     Comma-Separated Values
#
# A CSV file is a simple text-based format used to store
# tabular data, similar to spreadsheets or database tables.
#
# Each line represents one row.
# Values inside a row are separated using commas.
#
# Example:
#
# Name,City,Country
# John,Berlin,Germany
# Alice,Paris,France
#
# CSV files are very common in:
#   - data science
#   - machine learning
#   - databases
#   - data analysis pipelines


# -------------------------------------------------------------------
# Import the csv module
# -------------------------------------------------------------------

# Python provides a built-in csv module
# specifically designed for reading and writing CSV files.
import csv


# -------------------------------------------------------------------
# Open the CSV file
# -------------------------------------------------------------------

# Open the file named "csv_file.csv" in read mode.
#
# Since no mode is specified,
# Python automatically uses read mode ("r").
#
# The file object is stored in the variable f.
with open('csv_file.csv') as f:


    # ---------------------------------------------------------------
    # Create a CSV reader object
    # ---------------------------------------------------------------

    # csv.reader() reads the CSV file row by row.
    #
    # delimiter=','
    # means that commas separate the values.
    #
    # Each row is returned as a list.
    csv_reader = csv.reader(f, delimiter=',')


    # ---------------------------------------------------------------
    # Initialise a line counter
    # ---------------------------------------------------------------

    # line_count keeps track of how many rows
    # have been processed.
    #
    # It also helps us identify the header row.
    line_count = 0


    # ---------------------------------------------------------------
    # Iterate through each row in the CSV file
    # ---------------------------------------------------------------

    # The loop processes one row at a time.
    #
    # Each row is represented as a list.
    for row in csv_reader:


        # -----------------------------------------------------------
        # Process the first row (header row)
        # -----------------------------------------------------------

        # Usually, the first row in a CSV file
        # contains column names.
        if line_count == 0:


            # -------------------------------------------------------
            # Print the column names
            # -------------------------------------------------------

            # ", ".join(row)
            #
            # join() combines all list elements into one string,
            # separated by commas.
            #
            # **Dummy Example**:
            # ['Name', 'City', 'Country']
            #
            # becomes:
            # "Name, City, Country"
            print(f'Column names are :{", ".join(row)}')


            # Increase the line counter
            line_count += 1


        # -----------------------------------------------------------
        # Process the remaining rows (actual data)
        # -----------------------------------------------------------

        else:


            # -------------------------------------------------------
            # Access values using indexing
            # -------------------------------------------------------

            # row[0] → first column
            # row[1] → second column
            # row[2] → third column
            #
            # **Dummy Example**:
            # ['Tom Cruise', 'Los Angeles', 'USA']
            #
            # Output:
            # Tom Cruise is an Actor. He lives in Los Angeles, USA.
            print(
                f'{row[0]} is an Actor. He lives in {row[1]}, {row[2]}.'
            )


            # Increase the line counter
            line_count += 1


    # ---------------------------------------------------------------
    # Print the total number of processed lines
    # ---------------------------------------------------------------

    # This includes:
    #   - the header row
    #   - all data rows
    print(f'Number of lines:  {line_count}')


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    # Since we are using with open(),
    # Python automatically closes the file.
    f.close()
````

````python
# -------------------------------------------------------------------
# Creating a CSV file using dictionaries
# -------------------------------------------------------------------

# In this example:
#   - each actor record is stored as a dictionary
#   - dictionary keys represent column names
#   - dictionary values represent row data
#
# This approach is very common in:
#   - data science
#   - APIs
#   - JSON-to-CSV conversion
#   - structured data processing


# -------------------------------------------------------------------
# Import the csv module
# -------------------------------------------------------------------

# The built-in csv module provides tools
# for reading and writing CSV files.
import csv


# -------------------------------------------------------------------
# Step 1: Create the dataset
# -------------------------------------------------------------------

# data is a list of dictionaries.
#
# Each dictionary represents ONE row in the CSV file.
#
# Example:
# {
#     'firstname': 'Tom',
#     'lastname': 'Cruise',
#     'country': 'USA'
# }
#
# Dictionary structure:
#   key   → column name
#   value → actual row value
data = [
    {'firstname': 'Tom', 'lastname': 'Cruise', 'country': 'USA'},
    {'firstname': 'Tom', 'lastname': 'Hanks', 'country': 'USA'},
    {'firstname': 'Tom', 'lastname': 'Burke', 'country': 'UK'},
    {'firstname': 'Tom', 'lastname': 'Holland', 'country': 'UK'},
]


# -------------------------------------------------------------------
# Step 2: Open/Create the CSV file
# -------------------------------------------------------------------

# 'male_actor_records.csv'
#     → the name of the CSV file
#
# 'w'
#     → write mode
#
# Write mode:
#   - creates the file if it does not exist
#   - overwrites the file if it already exists
#
# newline=''
#
# Prevents extra empty lines from appearing
# in some operating systems such as Windows.
with open('male_actor_records.csv', 'w', newline='') as csvfile:


    # ---------------------------------------------------------------
    # Step 3: Define the CSV column names
    # ---------------------------------------------------------------

    # fieldnames defines the order of columns
    # inside the CSV file.
    #
    # The order matters because:
    # dictionaries themselves are not responsible
    # for controlling CSV column layout.
    fieldnames = ['firstname', 'lastname', 'country']


    # ---------------------------------------------------------------
    # Step 4: Create a DictWriter object
    # ---------------------------------------------------------------

    # csv.DictWriter() writes dictionaries into a CSV file.
    #
    # Parameters:
    #
    # csvfile
    #     → the opened file object
    #
    # fieldnames
    #     → defines the CSV column structure
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)


    # ---------------------------------------------------------------
    # Step 5: Write the CSV header
    # ---------------------------------------------------------------

    # writeheader() creates the first row of the CSV file.
    #
    # Output:
    # firstname,lastname,country
    writer.writeheader()


    # ---------------------------------------------------------------
    # Step 6: Write all rows into the CSV file
    # ---------------------------------------------------------------

    # writerows(data)
    #
    # Writes multiple dictionaries into the CSV file.
    #
    # Each dictionary becomes one row.
    writer.writerows(data)


    # ---------------------------------------------------------------
    # Step 7: Print confirmation message
    # ---------------------------------------------------------------

    print("A new file has been created via dictionaries")


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    # Since we are using with open(),
    # the file is automatically closed.
    csvfile.close()
    
````

````python
# -------------------------------------------------------------------
# Creating a CSV file using pipe delimiters
# -------------------------------------------------------------------

# By default, CSV files use commas (,) to separate values.
#
# Example:
# firstname,lastname,country
#
# However, CSV files can use other delimiters as well,
# depending on:
#   - system requirements
#   - database exports
#   - application compatibility
#
# In this example, we use:
#     |   (pipe delimiter)
#
# Example output:
# firstname|lastname|country
# Katie|Hall|UK


# -------------------------------------------------------------------
# Import the csv module
# -------------------------------------------------------------------

# The csv module provides tools
# for reading and writing CSV files.
import csv


# -------------------------------------------------------------------
# Step 1: Create the dataset
# -------------------------------------------------------------------

# data is a list of rows.
#
# Each inner list represents one row
# in the CSV file.
data = [

    # ---------------------------------------------------------------
    # Header row
    # ---------------------------------------------------------------

    ['firstname', 'lastname', 'country'],


    # ---------------------------------------------------------------
    # Data rows
    # ---------------------------------------------------------------

    ['Katie', 'Hall', 'UK'],
    ['Katie', 'Holmes', 'USA'],
    ['Katie', 'Douglas', 'Canada']
]


# -------------------------------------------------------------------
# Step 2: Open/Create the CSV file
# -------------------------------------------------------------------

# 'female_actor_records_pipe_delimiters.csv'
#     → output file name
#
# 'w'
#     → write mode
#
# newline=''
#     → prevents extra blank lines
#        on some operating systems
with open('female_actor_records_pipe_delimiters.csv', 'w', newline='') as csvfile:


    # ---------------------------------------------------------------
    # Step 3: Create a CSV writer object
    # ---------------------------------------------------------------

    # csv.writer() creates a writer object
    # responsible for writing rows into the file.
    #
    # delimiter='|'
    #
    # Changes the separator from:
    #     comma (,)
    #
    # to:
    #     pipe (|)
    #
    # Example row:
    # Katie|Hall|UK
    writer = csv.writer(csvfile, delimiter='|')


    # ---------------------------------------------------------------
    # Step 4: Write all rows into the CSV file
    # ---------------------------------------------------------------

    # writerows(data)
    #
    # Writes all rows from the list into the file.
    writer.writerows(data)


    # ---------------------------------------------------------------
    # Step 5: Print confirmation message
    # ---------------------------------------------------------------

    print("A new file has been created via csv.writer with Pipe Delimiters")


    # ---------------------------------------------------------------
    # Manual close() is unnecessary
    # ---------------------------------------------------------------

    # Since we are using with open(),
    # Python automatically closes the file.
    csvfile.close()
````

## **5. Error Handling and Exception Concepts**

### Concept
- Python can **catch runtime errors** using `try...except`.  
- Avoids crashing the program.

### Example

```python
try:
    number = int(input("Enter a number: "))
    print("Reciprocal:", 1/number)
except ValueError:
    print("Error: Not a valid number!")
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")
finally:
    print("Execution complete.")
```

````python
# -------------------------------------------------------------------
# Exception Handling in Python using try-except
# -------------------------------------------------------------------

# Exception handling allows Python programs
# to continue running even when an error occurs.
#
# Instead of crashing the program,
# Python can "catch" the error and handle it gracefully.
#
# Basic structure:
#
# try:
#     code that may produce an error
#
# except:
#     code executed if an error occurs


# -------------------------------------------------------------------
# Start the try block
# -------------------------------------------------------------------

# Python first executes everything inside the try block.
#
# If an error occurs at any point,
# Python immediately jumps to the except block.
try:


    # ---------------------------------------------------------------
    # Step 1: Ask the user to enter their name
    # ---------------------------------------------------------------

    # input() always returns data as a STRING.
    #
    # Example:
    # If the user enters:
    # Ahmed
    #
    # name becomes:
    # "Ahmed"
    name = input('Enter your name:')


    # ---------------------------------------------------------------
    # Step 2: Ask the user to enter their year of birth
    # ---------------------------------------------------------------

    # Even if the user enters numbers,
    # input() still returns a string.
    #
    # Example:
    # User enters:
    # 2000
    #
    # yob becomes:
    # "2000"
    yob = input('Enter your year of birth:')


    # ---------------------------------------------------------------
    # Step 3: Convert the year of birth into an integer
    # ---------------------------------------------------------------

    # int(yob)
    #
    # converts the string into an integer
    # so mathematical operations become possible.
    #
    # Example:
    # "2000" → 2000
    age = 2026 - int(yob)


    # ---------------------------------------------------------------
    # Step 4: Print the calculated age
    # ---------------------------------------------------------------

    print(f'Your name is {name} and you are {age} years old.')


    # ---------------------------------------------------------------
    # Step 5: Intentional error demonstration
    # ---------------------------------------------------------------

    # IMPORTANT:
    # yob is still a STRING variable.
    #
    # Example:
    # yob = "2000"
    #
    # Python CANNOT subtract:
    # integer - string
    #
    # This line will produce a TypeError.
    #
    # Error example:
    # TypeError:
    # unsupported operand type(s) for -: 'int' and 'str'
    age = 2026 - yob


    # ---------------------------------------------------------------
    # This line will NEVER execute
    # ---------------------------------------------------------------

    # Since the previous line produces an error,
    # Python immediately jumps to the except block.
    #
    # Therefore, this print statement is skipped.
    print(f'Your name is {name} and you are {age} years old.')


# -------------------------------------------------------------------
# Except block
# -------------------------------------------------------------------

# If ANY error occurs inside the try block,
# Python executes this block.
#
# Since a TypeError occurs above,
# this message will be printed.
except:


    # ---------------------------------------------------------------
    # Print an error message
    # ---------------------------------------------------------------

    print('Something went wrong')
````

````python
# Before running the code, try to predict what could go wrong.
#
# Examples of wrong inputs:
#   - Entering text instead of a number for the year of birth:
#       "two thousand"
#
#   - Entering the current year:
#       2026
#     This makes the age equal to 0 and causes a ZeroDivisionError.
#
#   - Entering a future year:
#       2030
#     This gives a negative age, which is logically incorrect,
#     although it does not necessarily crash the program.

try:
    # Ask the user to enter their name.
    #
    # input() always returns a string.
    name = input('Enter your name: ')

    # Ask the user to enter their year of birth.
    #
    # Even if the user enters a number,
    # Python stores it as a string at first.
    yob = input('Enter your year of birth: ')

    # Store the current year in a variable.
    current_year = 2026

    # Convert the year of birth from string to integer.
    #
    # This may raise a ValueError if the user enters
    # something that cannot be converted into an integer.
    age = current_year - int(yob)

    # If the calculated age is less than or equal to zero,
    # the input is problematic.
    if age <= 0:

        # This commented line would raise a TypeError
        # because Python cannot concatenate a string with None.
        #
        # name = name + None

        # This line raises a ZeroDivisionError when age is 0.
        result = 100 / age

    # Print the calculated age.
    print(f'Your name is {name} and you are {age} years old.')

    # Important:
    # result is only created inside the if block.
    #
    # If age > 0, the if block is skipped,
    # so result does not exist.
    #
    # This may cause a NameError.
    print(f'100 divided by your age is {result}')


# Handle TypeError errors.
#
# A TypeError occurs when an operation is applied
# to an incompatible data type.
except TypeError:
    print('Type error occurred: Cannot perform operations with incompatible types')


# Handle ValueError errors.
#
# A ValueError occurs here when int(yob) fails.
#
# Example:
# yob = "abc"
except ValueError:
    print('Value error occurred: Please enter a valid number for year of birth')


# Handle ZeroDivisionError errors.
#
# This occurs when age is 0.
#
# Example:
# yob = 2026
except ZeroDivisionError:
    print('Zero division error occurred: Age cannot be zero')


# The else block runs only if no exception occurs
# inside the try block.
else:
    print('No errors occurred during execution')


# The finally block always runs.
#
# It runs whether:
#   - an error occurs
#   - no error occurs
#   - the error is handled
finally:
    print('This block always executes, regardless of errors')
````

## **6. Basic Parsing Principles**

- Parsing = reading structured or semi-structured data and extracting useful information  
- Often used in **logs, CSV, JSON, or text files**

````python
# -------------------------------------------------------------------
# Example: Parsing a text file
# -------------------------------------------------------------------

# Parsing means:
# reading raw text data and converting it
# into a structured format that Python can use.
#
# In this example:
#   - student data is stored inside a text file
#   - each line contains:
#         Name,Grade
#
# Example:
# Alice,85
# Bob,90
#
# We will:
#   1. Create the text file
#   2. Read the file line by line
#   3. Split each line into separate values
#   4. Store the result inside a dictionary


# ===================================================================
# Step 1: Create and write into the text file
# ===================================================================

# Open the file in write mode ("w").
#
# If the file already exists:
#   → its old content will be overwritten
#
# If the file does not exist:
#   → Python creates it automatically
with open("grades.txt", "w") as f:


    # ---------------------------------------------------------------
    # Write student records into the file
    # ---------------------------------------------------------------

    # \n represents a newline character.
    #
    # Each student record is written on a separate line.
    #
    # Final file content:
    #
    # Alice,85
    # Bob,90
    # Charlie,78
    f.write("Alice,85\nBob,90\nCharlie,78\n")


# ===================================================================
# Step 2: Create an empty dictionary
# ===================================================================

# This dictionary will later store:
#
# key   → student name
# value → student grade
#
# Example:
# {
#     'Alice': 85,
#     'Bob': 90
# }
students = {}


# ===================================================================
# Step 3: Open the file in read mode
# ===================================================================

# Open the file again for reading.
with open("grades.txt", "r") as f:


    # ---------------------------------------------------------------
    # Read the file line by line
    # ---------------------------------------------------------------

    # The loop iterates through each line individually.
    #
    # Example first iteration:
    # line = "Alice,85\n"
    for line in f:


        # -----------------------------------------------------------
        # Remove newline characters and split the line
        # -----------------------------------------------------------

        # line.strip()
        #
        # Removes leading/trailing whitespace,
        # including newline characters (\n).
        #
        # Example:
        # "Alice,85\n"
        #
        # becomes:
        # "Alice,85"
        #
        #
        # .split(",")
        #
        # Splits the string into parts using commas.
        #
        # Example:
        # "Alice,85"
        #
        # becomes:
        # ['Alice', '85']
        #
        #
        # Multiple assignment:
        #
        # name  = 'Alice'
        # grade = '85'
        name, grade = line.strip().split(",")


        # -----------------------------------------------------------
        # Store data inside the dictionary
        # -----------------------------------------------------------

        # Convert grade from string to integer.
        #
        # Example:
        # "85" → 85
        #
        # Then store:
        # key   → student name
        # value → grade
        students[name] = int(grade)


# ===================================================================
# Step 4: Print the final dictionary
# ===================================================================

# Expected output:
#
# {
#     'Alice': 85,
#     'Bob': 90,
#     'Charlie': 78
# }
print(students)
````

````python
# -------------------------------------------------------------------
# Example: Basic CSV Parsing in Python
# -------------------------------------------------------------------

# Parsing means:
# reading structured data from a file
# and converting it into Python objects
# such as dictionaries or lists.
#
# In this example:
#   - we create a CSV file
#   - read it line by line
#   - parse each row
#   - store the result inside a list of dictionaries


# -------------------------------------------------------------------
# Import the csv module
# -------------------------------------------------------------------

# The csv module provides built-in support
# for reading and writing CSV files.
import csv


# ===================================================================
# Step 1: Create a sample CSV file
# ===================================================================

# Open/Create the CSV file in write mode.
with open("employees.csv", "w", newline='') as f:


    # ---------------------------------------------------------------
    # Create a CSV writer object
    # ---------------------------------------------------------------

    writer = csv.writer(f)


    # ---------------------------------------------------------------
    # Write the header row
    # ---------------------------------------------------------------

    # The first row usually contains column names.
    writer.writerow(["name", "department", "salary"])


    # ---------------------------------------------------------------
    # Write employee records
    # ---------------------------------------------------------------

    writer.writerow(["Alice", "Data Science", 5000])
    writer.writerow(["Bob", "Cyber Security", 4500])
    writer.writerow(["Charlie", "AI Engineering", 6000])


# ===================================================================
# Step 2: Read and parse the CSV file
# ===================================================================

# Create an empty list to store parsed records.
employees = []


# Open the CSV file in read mode.
with open("employees.csv", "r") as f:


    # ---------------------------------------------------------------
    # Create a CSV reader object
    # ---------------------------------------------------------------

    reader = csv.reader(f)


    # ---------------------------------------------------------------
    # Read the header row
    # ---------------------------------------------------------------

    # next(reader) reads the first row only.
    #
    # This row contains the column names.
    header = next(reader)

    print("Header:", header)


    # ---------------------------------------------------------------
    # Process the remaining rows
    # ---------------------------------------------------------------

    for row in reader:


        # -----------------------------------------------------------
        # Extract values from the row
        # -----------------------------------------------------------

        # Example row:
        # ['Alice', 'Data Science', '5000']
        #
        # row[0] → name
        # row[1] → department
        # row[2] → salary
        name = row[0]
        department = row[1]
        salary = int(row[2])


        # -----------------------------------------------------------
        # Create a dictionary for each employee
        # -----------------------------------------------------------

        employee_record = {
            "name": name,
            "department": department,
            "salary": salary
        }


        # -----------------------------------------------------------
        # Store the dictionary inside the list
        # -----------------------------------------------------------

        employees.append(employee_record)


# ===================================================================
# Step 3: Print the parsed data
# ===================================================================

# Final structured output:
#
# [
#     {'name': 'Alice', 'department': 'Data Science', 'salary': 5000},
#     {'name': 'Bob', 'department': 'Cyber Security', 'salary': 4500},
#     {'name': 'Charlie', 'department': 'AI Engineering', 'salary': 6000}
# ]
print(employees)
````

````python
# -------------------------------------------------------------------
# Example: Basic JSON Parsing in Python
# -------------------------------------------------------------------
# Import the json module
# -------------------------------------------------------------------

# The built-in json module allows Python
# to work with JSON files and JSON strings.
import json


# ===================================================================
# Step 1: Create sample JSON data
# ===================================================================

# Here, we create a Python dictionary
# containing employee information.
#
# The structure contains:
#   - a company name
#   - a list of employees
data = {
    "company": "AI Solutions",
    "employees": [
        {
            "name": "Alice",
            "department": "Data Science",
            "salary": 5000
        },
        {
            "name": "Bob",
            "department": "Cyber Security",
            "salary": 4500
        },
        {
            "name": "Charlie",
            "department": "AI Engineering",
            "salary": 6000
        }
    ]
}


# ===================================================================
# Step 2: Write the data into a JSON file
# ===================================================================

# Open/Create the JSON file in write mode.
with open("employees.json", "w") as f:


    # ---------------------------------------------------------------
    # Convert Python dictionary into JSON format
    # ---------------------------------------------------------------

    # json.dump()
    #
    # Writes Python objects into a JSON file.
    #
    # indent=4
    #
    # Makes the JSON file nicely formatted
    # and easier to read.
    json.dump(data, f, indent=4)


# ===================================================================
# Step 3: Read and parse the JSON file
# ===================================================================

# Open the JSON file in read mode.
with open("employees.json", "r") as f:


    # ---------------------------------------------------------------
    # Parse JSON into Python objects
    # ---------------------------------------------------------------

    # json.load()
    #
    # Reads JSON data from the file
    # and converts it into Python objects.
    #
    # JSON object  → Python dictionary
    # JSON array   → Python list
    parsed_data = json.load(f)


# ===================================================================
# Step 4: Access parsed JSON data
# ===================================================================

# Print the company name.
print("Company:", parsed_data["company"])


# -------------------------------------------------------------------
# Step 5: Access employee records
# -------------------------------------------------------------------

# parsed_data["employees"]
#
# returns a list of employee dictionaries.
for employee in parsed_data["employees"]:


    # ---------------------------------------------------------------
    # Extract employee information
    # ---------------------------------------------------------------

    name = employee["name"]
    department = employee["department"]
    salary = employee["salary"]


    # ---------------------------------------------------------------
    # Print employee details
    # ---------------------------------------------------------------

    print(
        f'{name} works in {department} '
        f'and earns {salary} euros.'
    )
````
