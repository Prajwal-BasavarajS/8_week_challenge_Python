```python
# np.array() converts a Python list into a NumPy array.
import numpy as np
sales_list= [200, 150,100, 140, 220, 180, 179]
sales = np.array(sales_list)
print(sales)
print ("Type of sales:", type(sales))
```

### Vectorised Operations 
Vectorisation in **NumPy** refers to applying operations on entire arrays without using explicit loops. 

```python
# computation of common descriptive statistics using vectorised NumPy operations

# total sum of all elements in the sales array
total = sales.sum()

# arithmetic mean of all elements in the sales array
mean = sales.mean()

# standard deviation of the sales distribution
std_dev = sales.std()

# output of computed summary statistics
print("Total:", total)
print("Mean:", mean)
print("Standard deviation:", std_dev)
```

```python
# Step-by-step computation of Standard Deviation **(for basic understanding purpose)**

# step 1: compute the mean
mean = sales.mean()

# step 2: compute deviations from the mean
deviations = sales - mean

# step 3: square the deviations
squared_deviations = deviations ** 2

# step 4: compute the variance (average of squared deviations)
variance = squared_deviations.mean()

# step 5: standard deviation is the square root of variance
std_dev = np.sqrt(variance)

print("Mean:", mean)
print("Deviations:", deviations)
print("Squared deviations:", squared_deviations)
print("Variance:", variance)
print("Standard deviation:", std_dev)
```

```python
# Aggregation Functions
A = np.array([[1, 2],  #2D Array
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Sum of all Elements
print("Sum of all elements in A", A.sum())
# Mean (Average)
print("Average of all elements in B", B.mean())
#  Standard deviation of all elements
print("Standard deviation of all elements in B", B.std())
# Min
print("Minimum element value in B", B.min())
# Max
print("Maximum element value in A", A.max())
```

### Element-wise Operations

With **NumPy**, you can perform math operations directly on arrays! **No need for loops**.

- Addition
- Subtraction
- Multiplication
- Division

```python
a = np.array([22, 23, 24])
b = np.array([2, 3, 4])

# Element-wise addition - 22+2, 23+3, 24+4
print(a + b)  

# Element-wise subtraction - 22-2, 23-3, 24-4
print(a-b) 

# Element-wise multiplication - 22*2, 23*3, 24*4
print(a * b)  

# Element-wise division - 2/22, 3/23, 4/2
print(b / a)  
```

```python
'''Look at the output of each operation'''
A = np.array([[1, 2],   
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Element-wise addition - 1+5 , 2+6
#                         3+7, 4+8
print("Element-wise addition",A+B)  

# Element-wise subtraction- 5-1 , 6-2
#                         7-3, 8-4
print("Element-wise subtraction",B-A) 

# Element-wise multiplication - 1*5 , 2*6
#                         3*7, 4*8
print("Element-wise multiplication",A*B)  

# Element-wise division 5/1 , 6/2
#                         7/3, 8/4
print("Element-wise division",B/A)  
```

```python
# Dot Product operation
"""The result is calculated by multiplying corresponding elements 
in a row of A with a column of B and summing them."""
""" [(1*5)+(2*7)  (1*6)+(2*8)
     (3*5)+(4*7)  (3*6)+(4*8)]"""

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Using np.dot()
print("np.dot(A, B)",np.dot(A, B))

# Or using the '@' operator (more Pythonic)
print("A @ B",A @ B)
```

```python
# Matrix Transpose
"""Flipping a matrix over its diagonal. Rows become columns, and columns become rows."""
Original_Matrix= np.array([[19, 22],
                           [43, 50]])
Transposed_Matrix= Original_Matrix.T
print(Transposed_Matrix)
```

This allows *NumPy* to perform operations on arrays of different shapes without making copies of data.  *NumPy*  automatically stretches the smaller array to match the shape of the larger one.

### Boolean Indexing
 With Boolean indexing, you can filter elements in an array based on conditions.

```python
# Working with Conditions in NumPy
one_arr = np.array([1, 2, 3, 4, 5, 6])

'''Get elements greater than 3
     one_arr > 3 creates a Boolean array and NumPy uses this to return only the elements where the condition is True '''
result = one_arr[one_arr > 3]
print(result)  
```

```python
# Working with Conditions in NumPy
two_arr = np.array([[12, 22, 32],
              [42, 52, 26]])

# Get elements greater than 12
result = two_arr[two_arr > 12]
print(result) # output in 1D
```

```python
# Working with multiple conditions in NumPy
two_arr = np.array([[12, 22, 32], [42, 52, 26]])

result = two_arr[(two_arr > 12) & (two_arr < 50)]
print(result)
```

```python
# Conditional Operations with where() - It’s like an if-else for arrays.
labeled_arr = np.array([100, 202, 150, 115, 160])

# Label elements as 'High' if >150, otherwise 'Low'
result = np.where(labeled_arr > 150, 'High', 'Low')
'''Breakdown:
Condition: labeled_arr > 150
If True, return 'High'
If False, return 'Low'
'''
print(result)  
```

```python
# Conditional Operations with where() - It’s like an if-else for arrays.
labeled_arr = np.array([100, 202, 150, 115, 160])

# replace elements greater than 150 with 250 , otherwise 0
result = np.where(labeled_arr > 150, 250, 0)
print(result)  
```

### Set Operations
Set operations in **NumPy** are just like that finding overlaps, differences, and unions.

```python
# unique()- No duplicates, just clean data
a = np.array([12, 22, 22, 23, 24, 24, 52])
print(np.unique(a))
```

```python
# intersect1d() - Find Common Elements
a = np.array([12, 22, 23, 42])
b = np.array([23, 42, 25, 62])
# Common elements
print(np.intersect1d(a, b)) 
```

```python
# union1d() - Combine Arrays (No Duplicates)
a = np.array([12, 22, 23, 42])
b = np.array([23, 42, 25, 62])
# union of both arrays
print(np.union1d(a, b)) 
```

```python
# setdiff1d() - Find Differences
a = np.array([12, 22, 23, 42])
b = np.array([23, 42, 25, 62])
# merges both arrays, removing duplicates.
print(np.setdiff1d(a, b)) 
print(np.setdiff1d(b, a)) 
```

```python
# array_split (arr, num)
a = np.array([12, 22, 23, 42])
newarr = np.array_split(a, 4)
print(newarr)
```

```python
# sort(arr) 
a = np.array([132, 22, 23, 42])
newarr = np.sort(a)
print(newarr)
```

```python
# Unique in 2d array
A = np.array([[1, 23],  
              [23, 4]])

B = np.array([[51, 22],
              [22, 8]])
print(np.unique(A))
print (np.unique(B))
```

```python
# Flatten the 2d array
A = np.array([[1, 23],  
              [23, 4]])
print(A.flatten())
```

```python
# Union in 2d arrays
A = np.array([[1, 23],  
              [23, 4]])

B = np.array([[51, 22],
              [22, 8]])
# To get the union of two arrays, use np.union1d
# First we need to flatten the arrays or not
print(np.union1d(A.flatten(), B.flatten()))
print(np.union1d(A,B))
```

```python
# intersection in 2d arrays
A = np.array([[1, 22],  
              [23, 4]])

B = np.array([[51, 22],
              [23, 8]])
# To get the intersection of two arrays, use np.intersect1d
# First we may flatten the arrays (optional)
print(np.intersect1d(A.flatten(), B.flatten()))
# or directly use intersect1d
print(np.intersect1d(A, B))
```

```python
# difference in 2d arrays  
A = np.array([[1, 22],  
              [23, 4]])

B = np.array([[51, 22],
              [23, 8]])
# Flatten the arrays and use np.unique to get unique values
# This is equivalent to a union operation for a single array
print(np.unique(A))
print(np.unique(B))

# To get the intersection of two arrays, use np.setdiff1d
# First we may flatten the arrays or not
print(np.setdiff1d(A.flatten(), B.flatten()))
print(np.setdiff1d(A, B))
print(np.setdiff1d(B.flatten(), A.flatten()))
print(np.setdiff1d(B, A))
```

```python
# reshape a NumPy array

A = np.array([[1, 22],  
              [23, 4]])

# reshape from 2x2 to 4x1
reshaped_A = np.reshape(A, (4, 1)) 
print(reshaped_A)
```

## Pandas Series

A Series is:

- One-dimensional

- Labelled

- Built on NumPy

```python
# pandas series syntax
import pandas as pd

'''First argument → data
   index → labels
'''

expenses = pd.Series(
     [450, 520, 610, 490, 580, 700],
    index=["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
   
)

print(expenses)
```

```python
# total expense
print("Total:", expenses.sum())
# average
print("Average:", expenses.mean())
# label of maximum value
print("Highest month:", expenses.idxmax())
```

```python
# Filtering a Series
high_expenses = expenses[expenses > 500] #Same Boolean logic as NumPy.
print(high_expenses)
```

## Grouping and Aggregation

```python
department_data = {
    "employee": ["A", "B", "C", "D", "E"],
    "department": ["IT", "HR", "IT", "HR", "Finance"],
    "salary": [4000, 3500, 4200, 3700, 5000]
}

df_dep = pd.DataFrame(department_data)
print (df_dep)
'''
- groupby("department") splits data.

- ["salary"] selects column.

- mean() aggregates.
'''
grouped = df_dep.groupby("department")["salary"].mean()
print(grouped)
```

## Data Cleaning

Real data is messy.

**Common problems**:

- Missing values

- Duplicates

- Wrong data types

- Inconsistent labels

```python
# Fill missing values
department_data = {
    "employee": ["A", "B", "C", "D", "E"],
    "department": ["IT", "HR", "IT", "HR", "Finance"],
    "salary": [4000, 3500, 4200, 3700, 5000]
}

df_dep = pd.DataFrame(department_data)
print(df_dep)

df_missing = df_dep.copy()

# Create a new row as a DataFrame
new_row = pd.DataFrame({"employee": ["F"], "department": ["Finance"], "salary": [None]})

# Use concat to merge old df with the new one
df_missing = pd.concat([df_missing, new_row], ignore_index=True)
pd.set_option('future.no_silent_downcasting', True)
   
print(df_missing)

# Fill missing values with mean
df_missing["salary"] = df_missing["salary"].fillna(df_missing["salary"].mean()).infer_objects(copy=False)
print(df_missing)
   
```

```python
department_data = {
    "employee": ["A", "B", "C", "D", "E"],
    "department": ["IT", "HR", "IT", "HR", "Finance"],
    "salary": [4000, 3500, 4200, 3700, 5000]
}

df_dep = pd.DataFrame(department_data)

df_missing = df_dep.copy()

new_row = pd.DataFrame({"employee": ["F"], "department": ["Finance"], "salary": 5000})
df_missing = pd.concat([df_missing, new_row], ignore_index=True)
df_missing = pd.concat([df_missing, new_row], ignore_index=True)
pd.set_option('future.no_silent_downcasting', True)
print(df_missing)

# Removing Duplicates
df_no_duplicates = df_missing.drop_duplicates()
print(df_no_duplicates)
```

```python
# Cleaning Strings - If salary contains currency symbol:
department_data = {
    "employee": ["A", "B", "C", "D", "E"],
    "department": ["IT", "HR", "IT", "HR", "Finance"],
    "salary": [4000, 3500, 4200, 3700, 5000]
}
# Create DataFrame (missing in original code)
import pandas as pd
df_dep = pd.DataFrame(department_data)

# Assign messy salary data
df_dep["salary"] = ["€4,000", "€3,500", "4200", " €3,700 ", "€5,000"]
print("\nMessy DataFrame:")
print(df_dep)

# Clean the salary column
df_dep["salary"] = df_dep["salary"].astype(str)
df_dep["salary"] = df_dep["salary"].str.replace("€", "")  # Remove euro symbol
df_dep["salary"] = df_dep["salary"].str.replace(",", "")  # Remove commas
df_dep["salary"] = df_dep["salary"].str.strip()  # Remove leading/trailing spaces
df_dep["salary"] = df_dep["salary"].astype(float)  # Convert to float

print("\nCleaned DataFrame:")
print(df_dep)
```
