# **Section A: Logical reasoning & comparison operators**

## **Problem A1 (Conditional number categorisation)**

Ask the user to enter a number. 

Print:
- "Positive even" if the number is positive and even
- "Positive odd" if positive and odd
- "Negative even" if negative and even
- "Negative odd" if negative and odd
- "Zero" if the number is 0

````python
x = int(input())

if x > 0 and x % 2 == 0:
    print("positive and even")
elif x > 0 and x % 2 != 0:
    print("positive and odd")
elif x < 0 and x % 2 == 0:
    print("negative and even")
elif x < 0 and x % 2 != 0:
    print("negative and odd")
else:
    print("Zero")
````

## **Problem A2 (Alphabet type)** 

Ask the user to enter a single character.

- "Consonant" if it is a consonant
- "Digit" if it is a number

````python
x = input("Enter a single character: ")

if x.lower() == 'a' or x.lower() == 'e' or x.lower() == 'i' or x.lower() == 'o' or x.lower() == 'u':
    print("Vowel")
elif x.lower() >= 'a' and x.lower() <= 'z':
    print("Consonant")
elif x >= '0' and x <= '9':
    print("Digit")
else:
    print("Other")


if x.lower() in "aeiou":
    print("vowel")
elif x.isalpha():
    print("consonant")
elif x.isdigit():
    print("digit")
else:
    print("others")


````

## **Problem A4 (Constraint-based validation)**

Ask the user to enter three integers: `a`, `b`, and `c`.

**Task:** print `Valid` if all the following conditions hold:
- `a < b < c`
- the difference between consecutive numbers is equal
- none of the numbers is negative

Otherwise, print `Invalid`.

### Reflection questions
- how does chained comparison differ from writing multiple separate comparisons?
- why is it useful to validate all constraints explicitly in one problem?


````python
a = int(input("enter a number"))
b = int(input("enter a number"))
c = int(input("enter a number"))

if a >= 0 and b >=0 and c>=0:
    if a<b and b<c:
        if (b-a) == (c-b):
            print("valid")
        else :
            print("invalid")
    else :
            print("invalid")
else :
            print("invalid")
````

## **Problem B1 (For loop with conditional accumulation)**

Ask the user to enter a positive integer `n`.

Using a `for` loop, calculate the sum of all numbers from `1` to `n` that are divisible by `3`.

Print the final sum.

### Reflection
- how can a loop be used to accumulate only selected values?
- what is the role of the conditional statement inside the loop?

````python
n = 9
ss = sum(numbers for numbers in range(0,n) if numbers % 3 == 0)
print(ss)
````

## **Problem B2 (For loop with counting categories)**

Ask the user to enter a word.

Using a `for` loop, count how many characters in the word are:
- vowels
- consonants

Assume the input contains only alphabetic characters.

Print both counts.

### Reflection
- how does a loop help process one character at a time?
- why is it useful to normalise text before checking conditions?


````python
x = input("enter a word")
cv = 0
cb = 0
for i in x:
    if i.lower() in "aeiou":
        cv += 1
    elif i.isalpha():
        cb += 1

print("vowels count",cv)
print("consonants count",cb)
    
    
````

## **Problem B4  (Loop control statements: continue and break)**

Ask the user to enter a positive integer `n`.

Using a `for` loop:
- skip numbers divisible by `4`
- stop the loop completely if the number reaches `13`
- print all other numbers from `1` to `n`

### Reflection
- what is the difference between `continue` and `break`?
- how does each statement affect the flow of the loop?

````python
x = int(input("enter a number"))
for i in range(x+1):
    if i % 4 == 0 :
        continue
    elif i == 13 :
        break
    else:
        print(i)
        
````

## **Problem B5 (While loop with input validation)**

Keep asking the user to enter a number until they enter a negative number.

Using a `while` loop:
- count how many non-negative numbers were entered
- compute their total sum

When the user enters a negative number, stop the loop and print:
- the count
- the sum

### Reflection
- why is a `while` loop more suitable here than a `for` loop?
- what condition controls the repetition in this task?

````python
x = int(input("Enter a number "))
count , total = 0 , 0
while(x>0):
    count += 1
    total += x 
    x = int(input("Enter a -ve number to stop"))

print("Count:", count)
print("Sum:", total)
````

# Final Mini Challenge: Smart Student Performance Analyser 
You are given a dataset of students, where each student has a name, a score, and an attendance percentage.
### Given Data
```
students = [
    {"name": "Ali", "score": 85, "attendance": 92},
    {"name": "Sara", "score": 58, "attendance": 75},
    {"name": "Lina", "score": 73, "attendance": 88},
    {"name": "Omar", "score": 49, "attendance": 95},
    {"name": "Noor", "score": 91, "attendance": 98},
    {"name": "Yara", "score": 66, "attendance": 60}
]
```
### Tasks

#### Part 1: Classify each student
- if score ≥ 70 and attendance ≥ 80 → "Good standing"  
- if score < 70 and attendance ≥ 80 → "Needs academic improvement"  
- if score ≥ 70 and attendance < 80 → "Needs attendance improvement"  
- otherwise → "At risk"  

#### Part 2: Count Categories
Print the total number of students in each category

#### Part 3: Warning System
Print warnings if:
- student is "At risk"  
- score < 50  
- attendance < 65  

#### Part 4: List Comprehension
Create a list of students in "Good standing"



````python
students = [
    {"name": "Ali", "score": 85, "attendance": 92},
    {"name": "Sara", "score": 58, "attendance": 75},
    {"name": "Lina", "score": 73, "attendance": 88},
    {"name": "Omar", "score": 49, "attendance": 95},
    {"name": "Noor", "score": 91, "attendance": 98},
    {"name": "Yara", "score": 66, "attendance": 60}
]

# Part 2: Count categories
counts = {
    "Good standing": 0,
    "Needs academic improvement": 0,
    "Needs attendance improvement": 0,
    "At risk": 0
}

print("Student Classification")

for student in students:
    score = student["score"]
    attendance = student["attendance"]

    # Part 1: Classify each student
    if score >= 70 and attendance >= 80:
        category = "Good standing"
    elif score < 70 and attendance >= 80:
        category = "Needs academic improvement"
    elif score >= 70 and attendance < 80:
        category = "Needs attendance improvement"
    else:
        category = "At risk"

    print(student["name"], "->", category)

    # Count categories
    counts[category] += 1

print("\nCategory Counts")
for category, count in counts.items():
    print(category + ":", count)

# Part 3: Warning System
print("\nWarnings")
for student in students:
    score = student["score"]
    attendance = student["attendance"]

    if score >= 70 and attendance >= 80:
        category = "Good standing"
    elif score < 70 and attendance >= 80:
        category = "Needs academic improvement"
    elif score >= 70 and attendance < 80:
        category = "Needs attendance improvement"
    else:
        category = "At risk"

    if category == "At risk":
        print(student["name"], "- At risk")

    if score < 50:
        print(student["name"], "- Score below 50")

    if attendance < 65:
        print(student["name"], "- Attendance below 65%")

# Part 4: List Comprehension
good_students = [
    student["name"]
    for student in students
    if student["score"] >= 70 and student["attendance"] >= 80
]

print("\nStudents in Good Standing:")
print(good_students)
````
